```python
import os
import glob
from multiprocessing import Pool
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
def main():
    stock_path_list = glob.glob("./stock_indicator2018/nikkei/*")
    stock_df = import_file(stock_path_list)
    make_graph(stock_df)
```

```python
def import_file(csv_path_list):
    p = Pool(os.cpu_count())
    df = pd.concat(p.map(multi_read, csv_path_list), axis=0)
    p.close()
    return df
```

```python
def multi_read(csv_path):
    df = pd.read_csv(csv_path)
    df['PSAR'] = df.PSARBULL.fillna(0) - df.PSARBEAR.fillna(0)
    df = df.drop(['PSARBULL', 'PSARBEAR'], axis=1)
    scalar = StandardScaler()
    date = df[['value_date', 'stock_code']]
    value = pd.DataFrame(scalar.fit_transform(df.loc[:, 'open':]))
    df_std = pd.concat([date, value], axis=1,  join_axes=[date.index])
    df_std.columns = df.columns
    return df_std
```

```python
def make_graph(stock_df):
    file_name = 'graph.gsp'
    f = open(file_name, mode='w')
    for date in stock_df.value_date.unique():
        # TODO select indicator
        daily_df = stock_df[stock_df.value_date == date].drop(['value_date', 'stock_code'], axis=1)
        daily_coef = np.corrcoef(daily_df)
        if not np.isnan(daily_coef).all():
            # TODO make graph
            f.write(f't # 0 {date}\n')
            for i, node in enumerate(stock_df.stock_code.unique()):
                f.write(f'v {i} {node}\n')
            for i in range(len(daily_coef)):
                for j in range(i+1, len(daily_coef)):
                    if daily_coef[i][j] >= 0.7:
                        f.write(f'e {i} {j} 0\n')
            f.write('\n')
```

```python
if __name__ == '__main__':
    main()
```
